```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
String_spacing = {'IceCube':125,'IceCube-Gen2':240,'ANTARES':70,'KM3NeT/ORCA':20,'KM3NeT/ARCA':90,'Baikal/GVD':60}
OM_spacing = {'IceCube':17,'IceCube-Gen2':17,'ANTARES':14.5,'KM3NeT/ORCA':9,'KM3NeT/ARCA':26,'Baikal/GVD':15}
```

```python
fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(1,1,1)
ax.tick_params(which='major',length=5,width=1)
ax.tick_params(which='minor',length=5,width=1)
ax.tick_params(which='both')#,direction='in')
ax.tick_params(labelsize=15)

markersize_value=10
ax.plot(String_spacing['IceCube'],OM_spacing['IceCube'],'P',markersize=markersize_value,label='IceCube',color='blue')
ax.plot(String_spacing['IceCube-Gen2'],OM_spacing['IceCube-Gen2'],'P',markersize=markersize_value,label='IceCube-Gen2',color='blue',fillstyle='none')
ax.plot(String_spacing['ANTARES'],OM_spacing['ANTARES'],'^',markersize=markersize_value,label='ANTARES',color='fuchsia')
ax.plot(String_spacing['KM3NeT/ORCA'],OM_spacing['KM3NeT/ORCA'],'*',markersize=markersize_value,label='KM3NeT/ORCA',color='red',fillstyle='none')
ax.plot(String_spacing['KM3NeT/ARCA'],OM_spacing['KM3NeT/ARCA'],'*',markersize=markersize_value,label='KM3NeT/ARCA',color='red')
ax.plot(String_spacing['Baikal/GVD'],OM_spacing['Baikal/GVD'],'d',markersize=markersize_value,label='Baikal/GVD',color='black')
#ax.legend(loc = 'upper right', fontsize=13)
font_value=13
annotation_size = 9
ax.text(129,17.5,'IceCube',color='blue',fontsize=font_value)
ax.text(129,18.5,'(~ 0.9Gton, 5160 OMs)',color='blue',fontsize=annotation_size)
ax.text(190,17.5,'IceCube-Gen2',color='blue',fontsize=font_value)
ax.text(190,18.5,'(~ 7.1Gton, 9600 OMs)',color='blue',fontsize=annotation_size)
ax.text(70,13.3,'ANTARES',color='fuchsia',fontsize=font_value)
ax.text(70,12.5,'(~ 30Mton, 295 OMs)',color='fuchsia',fontsize=annotation_size)
ax.text(24,9.3,'KM3NeT/ORCA',color='red',fontsize=font_value)
ax.text(24,10.3,'(~ 6Mton, 2070 OMs)',color='red',fontsize=annotation_size)
ax.text(95,25.5,'KM3NeT/ARCA',color='red',fontsize=font_value)
ax.text(95,24.8,'(~ 1Gton, 4140 OMs)',color='red',fontsize=annotation_size)
ax.text(57,15.7,'Baikal/GVD',color='black',fontsize=font_value)
ax.text(57,16.7,'(~ 400Mton, 2304 OMs)',color='black',fontsize=annotation_size)


ax.set_xlabel('String Spacing (m)',size=18,labelpad=15)
ax.set_ylabel('Optical Module Spacing (m)',size=18,labelpad=15)
ax.set_xlim(0,250)
ax.set_ylim(7.5,27.5)
plt.savefig('./String-OM.pdf',bbox_inches = 'tight')
```
